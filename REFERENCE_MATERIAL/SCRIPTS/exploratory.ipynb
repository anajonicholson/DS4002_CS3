# Exploratory Data Analysis

### Reading in Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


personal_healthcare = pd.read_csv("https://raw.githubusercontent.com/anajonicholson/DS4002-Project2/refs/heads/main/DATA/personal_healthcare_expenses.csv")
print(personal_healthcare.shape)
personal_healthcare.head()
```

```python
medicaid_enrollment = pd.read_csv("https://raw.githubusercontent.com/anajonicholson/DS4002-Project2/refs/heads/main/DATA/medicaid_enrollment.csv")
print(medicaid_enrollment.shape)
medicaid_enrollment.head()
```

```python
state_population = pd.read_csv("https://github.com/anajonicholson/DS4002-Project2/blob/main/DATA/state_population.csv")
print(state_population.shape)
state_population.head()
```

### Data Cleaning
State Population Dataframe

```python
# Dropping first second two columns
state_population.drop(columns=['CENSUS2010POP', 'ESTIMATESBASE2010'], inplace=True)

# Rename Name column to State
state_population = state_population.rename(columns={'NAME': 'State'})

state_population.head()
```

```python
print(state_population.columns)
```

```python
# Only want data from 2014 to 2020
state_population = state_population[['State', 'POPESTIMATE2014', 'POPESTIMATE2015', 'POPESTIMATE2016', 'POPESTIMATE2017', 'POPESTIMATE2018', 'POPESTIMATE2019', 'POPESTIMATE2020']]

# Rename year columns
new_columns = ['State'] + [str(year) for year in range(2014, 2021)]
state_population.columns = new_columns

state_population.head()
```

```python
# Checking for missing values
print(state_population.isnull().sum())
```

```python
# Data types of all columns
print(state_population.dtypes)
```

Personal Healthcare Expense Dataframe

```python
# Drop the first row and set second row as the header
personal_healthcare.columns = personal_healthcare.iloc[0]
personal_healthcare = personal_healthcare[1:]
personal_healthcare.reset_index(drop=True, inplace=True)

personal_healthcare.head()
```

```python
# Dropping columns for year 1991-2013
columns_to_drop = [str(year) for year in range(1991, 2014)]
personal_healthcare.drop(columns=columns_to_drop, inplace=True)

# Drop Average Annual Growth Columns
personal_healthcare.drop(columns="Average\nAnnual %\nGrowth\n(1991-2020)", inplace=True)

# Dropping rows for regions and US (only want states)
regions = ["United States", "New England", "Mideast", "Great Lakes", "Plains", "Southeast", "Southwest", "Rocky Mountains", "Far West"]
personal_healthcare = personal_healthcare[~personal_healthcare['Region/state of residence'].isin(regions)]
personal_healthcare.reset_index(drop=True, inplace=True)

# Rename Region/state of residence column
personal_healthcare = personal_healthcare.rename(columns={'Region/state of residence': 'State'})

personal_healthcare.head()
```

```python
# Checking for missing values
print(personal_healthcare.isnull().sum())
```

```python
# Display the entire DataFrame
print(personal_healthcare.to_string())
```

```python
# Remove bottom three rows
personal_healthcare = personal_healthcare.iloc[:-3]

print(personal_healthcare.isnull().sum())
print(personal_healthcare.to_string())
```

```python
# Data types of all columns
print(personal_healthcare.dtypes)
```

```python
# Remove commas and convert to integers for all year columns
year_columns = personal_healthcare.columns[1:]

for col in year_columns:
    personal_healthcare[col] = personal_healthcare[col].replace(',', '', regex=True).astype(int)

print(personal_healthcare.dtypes)
personal_healthcare.head()
```

Medicaid Enrollment Dataframe

```python
# Drop the first row and set second row as the header
medicaid_enrollment.columns = medicaid_enrollment.iloc[0]
medicaid_enrollment = medicaid_enrollment[1:]
medicaid_enrollment.reset_index(drop=True, inplace=True)

medicaid_enrollment.head()
```

```python
# Dropping columns for year 1991-2013
columns_to_drop = [str(year) for year in range(1991, 2014)]
medicaid_enrollment.drop(columns=columns_to_drop, inplace=True)

# Drop Average Annual Growth Columns
medicaid_enrollment.drop(columns="Average\nAnnual %\nGrowth\n(1991-2020)", inplace=True)

# Dropping rows for regions and US (only want states)
regions = ["United States", "New England", "Mideast", "Great Lakes", "Plains", "Southeast", "Southwest", "Rocky Mountains", "Far West"]
medicaid_enrollment = medicaid_enrollment[~medicaid_enrollment['Region/state of residence'].isin(regions)]
medicaid_enrollment.reset_index(drop=True, inplace=True)

# Rename Region/state of residence column
medicaid_enrollment = medicaid_enrollment.rename(columns={'Region/state of residence': 'State'})

medicaid_enrollment.head()
```

```python
# Checking for missing values
print(medicaid_enrollment.isnull().sum())
```

```python
# Display the entire DataFrame
print(medicaid_enrollment.to_string())
```

```python
# Remove bottom three rows
medicaid_enrollment = medicaid_enrollment.iloc[:-3]

print(medicaid_enrollment.isnull().sum())
print(medicaid_enrollment.to_string())
```

```python
# Data types of all columns
print(medicaid_enrollment.dtypes)
```

```python
# Remove commas and convert to integers for all year columns
year_columns = medicaid_enrollment.columns[1:]

for col in year_columns:
    medicaid_enrollment[col] = medicaid_enrollment[col].replace(',', '', regex=True).astype(int)

print(medicaid_enrollment.dtypes)
medicaid_enrollment.head()
```

Creating combined dataframe

```python
print(state_population.columns)
print(medicaid_enrollment.columns)
```

```python
# Joining state_population and medicaid_enrollment dataframes
combined_data = pd.merge(state_population, medicaid_enrollment, on='State', suffixes=('_population', '_medicaid'))

# Calculate the percentage of the population enrolled in Medicaid for each year
for year in range(2014, 2021):
    population_col = f"{year}_population"
    medicaid_col = f"{year}_medicaid"
    percentage_col = f"{year}_percent_enrolled"

    combined_data[percentage_col] = (combined_data[medicaid_col] / combined_data[population_col]) * 100

# Select the 'State' and percentage columns
percent_columns = ['State'] + [f"{year}_percent_enrolled" for year in range(2014, 2021)]
medicaid_enrollment_percentage = combined_data[percent_columns]

# Rename columns
new_column_names = ['State'] + [str(year) for year in range(2014, 2021)]
medicaid_enrollment_percentage.columns = new_column_names

medicaid_enrollment_percentage.head()
```

### Exploratory Plots

```python
# Calculate the average personal healthcare expenses across the 7 years for each state
year_columns = personal_healthcare.columns[1:]
personal_healthcare['Average_Expense'] = personal_healthcare[year_columns].mean(axis=1)

# Creating bar plot
plt.figure(figsize=(10, 6))
plt.bar(personal_healthcare['State'], personal_healthcare['Average_Expense'])

plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Average Healthcare Expense (2014-2020)')
plt.title('Average Total Personal Healthcare Expenses by State (2014-2020)')
plt.tight_layout()

plt.show()
```

```python
# Calculate the average Medicaid enrollment across the 7 years for each state
year_columns_medicaid = medicaid_enrollment.columns[1:]
medicaid_enrollment['Average_Enrollment'] = medicaid_enrollment[year_columns_medicaid].mean(axis=1)

# Create a bar plot for Medicaid enrollment
plt.figure(figsize=(10, 6))
plt.bar(medicaid_enrollment['State'], medicaid_enrollment['Average_Enrollment'])

# Customize the plot
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Average Medicaid Enrollment (2014-2020)')
plt.title('Average Medicaid Enrollment by State (2014-2020)')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Reshape the DataFrame to a long format suitable for faceted plotting
personal_healthcare_long = pd.melt(personal_healthcare,
                                   id_vars=['State'],
                                   value_vars=['2014', '2015', '2016', '2017', '2018', '2019', '2020'],
                                   var_name='Year',
                                   value_name='Expense')

# Convert 'Year' to an integer type for proper sorting
personal_healthcare_long['Year'] = personal_healthcare_long['Year'].astype(int)
```

```python
g = sns.FacetGrid(personal_healthcare_long, col='State', col_wrap=3, height=3, sharey=False, despine=False)
g.map(sns.lineplot, 'Year', 'Expense')
g.set_titles("{col_name}")
g.set_axis_labels("Year", "Personal Healthcare Expense")

# Set x-ticks for all subplots without displaying the labels
for ax in g.axes.flatten():
    ax.set_xticks([2014, 2015, 2016, 2017, 2018, 2019, 2020])
    ax.set_xticklabels([])  # Hide the labels, keep only the ticks

plt.suptitle("Personal Healthcare Expenses Per Year by State (2014-2020)", fontsize=12, y=1.01)
plt.show()
```

```python
# Maryland and Georgia data (Adaopted medicaid expansion in 2014 vs Hasn't adopted medicaid expansion)
md_ga_data = personal_healthcare[personal_healthcare['State'].isin(['Maryland', 'Georgia'])]

# Reshape to long format
md_ga_long = pd.melt(md_ga_data, id_vars=['State'], value_vars=['2014', '2015', '2016', '2017', '2018', '2019', '2020'],
                     var_name='Year', value_name='Expense')

# Converting Year to int
md_ga_long['Year'] = md_ga_long['Year'].astype(int)

# Creating plot comparing Maryland and Georgia
plt.figure(figsize=(8, 6))
sns.lineplot(data=md_ga_long, x='Year', y='Expense', hue='State', marker='o')

plt.xlabel('Year')
plt.ylabel('Personal Healthcare Expense')
plt.title('Yearly Personal Healthcare Expense Comparison: Maryland vs. Georgia')
plt.legend(title='State')
plt.tight_layout()

plt.show()
```
